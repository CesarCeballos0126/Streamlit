```python
# Récuperer les données
df = pd.read_csv('eco2mix-regional-cons-def.csv', parse_dates=['Date'],sep=';');
```

```python
# Choisir la frequence de prediction, groupage 
periods = dict([('W',52),('D',365),('S',4),('M',12), ('Y',1)])
PeriodName = 'D'
PeriodValue = periods[PeriodName]
```

```python
# suppression des champs inutiles.
df  = df[['Code INSEE région', 'Région', 'Nature', 'Date', 
       'Consommation (MW)', 'Thermique (MW)', 'Nucléaire (MW)',
       'Eolien (MW)', 'Solaire (MW)', 'Hydraulique (MW)', 'Pompage (MW)',
       'Bioénergies (MW)']]
# Supprimer les lignes avec des valeurs absents
df.dropna(axis = 0, subset=["Consommation (MW)"], inplace=True)

# remplacer les Nan = 0 car pas production
df["Eolien (MW)"]= df["Eolien (MW)"].fillna(0)
df["Nucléaire (MW)"]= df["Nucléaire (MW)"].fillna(0)
df["Pompage (MW)"]= df["Pompage (MW)"].fillna(0)
```

```python
# BD pour de traitements globals
df_group = df

df_group = df_group.groupby(["Code INSEE région", 'Région'], as_index=False).agg({'Consommation (MW)':'mean', 'Thermique (MW)':'mean',
 'Nucléaire (MW)':'mean','Eolien (MW)':'mean', 'Solaire (MW)':'mean', 'Hydraulique (MW)':'mean', 'Pompage (MW)':'mean','Bioénergies (MW)':'mean'})
```

```python
# BD pour de traitemens par date
df = pd.DataFrame(df.set_index('Date').groupby(['Code INSEE région','Région'])['Consommation (MW)', 'Thermique (MW)', 'Nucléaire (MW)',
       'Eolien (MW)', 'Solaire (MW)', 'Hydraulique (MW)', 'Pompage (MW)','Bioénergies (MW)'].resample(PeriodName).mean())
```
